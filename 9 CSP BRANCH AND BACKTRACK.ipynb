BACKTRACK


```python
class nQueens:
    def __init__(self, n):
        self.n = n
        self.board = [[0 for x in range(n)] for y in range(n)]
        self.solutions = []

    def is_safe(self, row, col):
        for i in range(row):
            if self.board[i][col] == 1:
                return False
            for j in range(self.n):
                if (self.board[i][j] == 1 and
                   (abs(i - row) == abs(j - col))):
                    return False
        return True

    def solve(self, row):
        if row == self.n:
            solution = []
            for i in range(self.n):
                row_solution = ''
                for j in range(self.n):
                    if self.board[i][j] == 1:
                        row_solution += 'Q'
                    else:
                        row_solution += '.'
                solution.append(row_solution)
            self.solutions.append(solution)
            return

        for col in range(self.n):
            if self.is_safe(row, col):
                self.board[row][col] = 1
                self.solve(row + 1)
                self.board[row][col] = 0  # Backtrack

    def print_solutions(self):
        if not self.solutions:
            print("No solutions found.")
        else:
            print(f"\nTotal solutions: {len(self.solutions)}\n")
            for solution in self.solutions:
                for row in solution:
                    print(row)
                print()


# Main Execution
if __name__ == '__main__':
    print("Name:Prashant Bankar.\nRoll No.:TACO22153")
    while(True):
        
        n = int(input("Enter the size of the chessboard: "))
        if (n%2==0):
            queens = nQueens(n)
            queens.solve(0)
            queens.print_solutions()
            break
        else:
            print("Please enter a valid number in the power of 2")
    
```

BRANCH AND BOUND

```python
def is_safe(arr, x, y, n):
    for row in range(x):
        if arr[row][y] == 1:
            return False

    row, col = x, y
    while row >= 0 and col >= 0:
        if arr[row][col] == 1:
            return False
        row -= 1
        col -= 1

    row, col = x, y
    while row >= 0 and col < n:
        if arr[row][col] == 1:
            return False
        row -= 1
        col += 1

    return True

def branch_and_bound_nqueen(arr, x, n):
    if x >= n:
        return True

    for col in range(n):
        if is_safe(arr, x, col, n):
            arr[x][col] = 1
            if branch_and_bound_nqueen(arr, x + 1, n):
                return True
            arr[x][col] = 0
    return False

def main():
    print("Name: PRASHANT BANKAR")
    print("Roll no: TACO22153")
    n = int(input("Enter the number of Queens: "))
    arr = [[0] * n for _ in range(n)]
    if branch_and_bound_nqueen(arr, 0, n):
        for row in arr:
            print(' '.join(str(cell) for cell in row))
    else:
        print("No solution found.")

if __name__ == '__main__':
    main()
```

```python
steps = 0

def print_board(board):
    for row in board:
        print(' '.join(row))
    print()

def solve(row, n, board, cols, diag1, diag2):
    global steps
    steps += 1
    if row == n:
        print("Final Solution Found:")
        print_board(board)
        return True
    for col in range(n):
        if not cols[col] and not diag1[row - col] and not diag2[row + col]:
            board[row][col] = 'Q'
            cols[col] = diag1[row - col] = diag2[row + col] = True
            print(f"Step {steps}: Placing Q at ({row}, {col})")
            print_board(board)
            if solve(row + 1, n, board, cols, diag1, diag2):
                return True
            # Undo placement (backtrack silently as per Branch and Bound)
            board[row][col] = '.'
            cols[col] = diag1[row - col] = diag2[row + col] = False
    return False

def main():
    global steps
    print("Name: PRASHANT BANKAR")
    print("Roll No: TACO22153")
    n = int(input("Enter the number of Queens: "))
    board = [['.'] * n for _ in range(n)]
    cols = [False] * n
    diag1 = {}
    diag2 = {}
    for i in range(-n, n): diag1[i] = False
    for i in range(2 * n): diag2[i] = False

    if not solve(0, n, board, cols, diag1, diag2):
        print("No solution found.")
    print("Total steps:", steps)

main()
```
